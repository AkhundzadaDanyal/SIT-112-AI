### B. Write Python code that performs (answers) the following operations (questions)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

##### 1.	Read the data from the CSV file into a DataFrame and display the first five rows.

```python
# Load the avocado data using pandas
avocado_data = pd.read_csv('avocado.csv')

# Display the first few rows of the dataset
avocado_data.head()
```

##### 2.	Display basic information about the DataFrame and its columns to see that the Date column has the object type.

```python
# Display information about the avocado dataset
avocado_data.info()
```

##### 3.	Convert the Date column to the datetime type.

```python
# Convert the 'Date' column to datetime format
avocado_data['Date'] = pd.to_datetime(avocado_data['Date'])
```

##### 4.	Filter the data so it contains just the rows for 2015, for the conventional type, and for the region named Albany. Store just the Date, Total Bags, and Small Bags columns in a new DataFrame.

```python
# Filter the avocado data for the year 2015, the region "Albany", and the type "conventional"
filtered_data = avocado_data[(avocado_data['year'] == 2015) & (avocado_data['region'] == "Albany") & (avocado_data['type'] == "conventional")]

# Select only the columns "Date", "Total Bags", and "Small Bags" from the filtered data
selected_data = filtered_data[['Date', 'Total Bags', 'Small Bags']]
```

##### 5.	Melt the data in the Total Bags and Small Bags columns, but not the values in the Date column. Name the column that contains the type of bag Bags, and name the column that contains the number of bags Count. Then, display the resulting DataFrame.

```python
# Reshape the avocado data using pd.melt(), specifying "Date" as the identifier variable
# and "Total Bags" and "Small Bags" columns as the value variables
melted_data = pd.melt(filtered_data, id_vars='Date', value_vars=['Total Bags', 'Small Bags'],
                      var_name='Bags', value_name='Count')

# Display the reshaped data
melted_data
```

##### 6.	Plot the melted data with Seaborn in a line plot, using the hue parameter to distinguish between the bag types.

```python
# Plot the data using seaborn's relplot, specifying line plot type, and customizing color
sns.relplot(data=melted_data, kind='line', x='Date', y='Count', hue='Bags', palette=['blue', 'green'])

# Display the plot
plt.show()
```

##### 7.	Bin the data in the Total Volume column into four quantiles labeled ‘poor’, ‘modest’, ‘good’, and ‘excellent’, and store the bin labels in a new column.

```python
# Create a new column 'Sales Volume' based on quartiles of 'Total Volume'
avocado_data['Sales Volume'] = pd.qcut(avocado_data['Total Volume'], q=4, labels=['poor', 'modest', 'good', 'excellent'])

# Display the updated dataset
avocado_data
```

##### 8.	Plot the binned data by year using a Seaborn count plot.

```python
# Plot the count of sales volume categories over the years using seaborn's catplot,
# specifying bar plot type, and customizing color
sns.catplot(data=avocado_data, kind='count', x='year', hue='Sales Volume', palette='Set2')

# Display the plot
plt.show()
```

##### 9.	Of the three bag sizes, which size sells the most? Hint: Melt the three bag size columns and plot the bags by year.

```python
# Melt the data for plotting by year and bags
melted_data_2 = pd.melt(avocado_data, id_vars='year', value_vars=['Small Bags', 'Large Bags', 'XLarge Bags'],
                        var_name='Bags', value_name='Count')

# Plot the data using seaborn's relplot, specifying line plot type, and customizing color
sns.relplot(data=melted_data_2, kind='line', x='year', y='Count', hue='Bags', palette=['blue', 'green', 'orange'], errorbar=None)

# Display the plot
plt.show()
```

##### 10.	Which type of avocado sells the most?

```python
# Group the avocado data by type and calculate the sum of 'Total Volume'
total_volume_by_type = avocado_data.groupby('type')['Total Volume'].sum()

# Display the total volume by type
total_volume_by_type
```
