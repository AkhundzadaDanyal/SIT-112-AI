```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### B. Write Python code that performs (answers) the following operations (questions) 

##### 1. Read the data from the .csv file into a DataFrame and display the first five rows.

```python
# Read the avocado dataset from a CSV file
data = pd.read_csv('avocado.csv')

# Display the first few rows of the dataset to get an overview of the data structure
print(data.head())
```

##### 2. Filter the data to exclude the following regions: TotalUS, West, and NorthEast. Use the filtered data to create a DataFrame that contains the total of the Small Bags, Large Bags, and XLarge Bags columns grouped by type. Use the grouped data to create a bar plot that shows the number of small, large, and extra-large bags for both types of avocado. Set the tile and axis labels of the bar plot and make sure all the labels are readable. 

```python
# Filter the data to exclude regions that are not of interest
regions_to_exclude = ['TotalUS', 'West', 'NorthEast']
avocado_filtered = data[~data['region'].isin(regions_to_exclude)]

# Aggregate the data by avocado type to calculate the total number of small, large, and extra-large bags
bags_totals = avocado_filtered.groupby('type').sum()[['Small Bags', 'Large Bags', 'XLarge Bags']]

# Create a bar plot using the aggregated data
plot = bags_totals.plot(kind='bar')

# Set the plot title and axis labels with appropriate spacing
plot.set_title('Total Number of Small, Large, and XLarge Bags by Avocado Type', pad=10)
plot.set_xlabel('Avocado Type')
plot.set_ylabel('Number of Bags')

# Display the plot
plt.show()

# Alternative approach using seaborn
# Reset the index of the grouped data to prepare it for melting
bags_totals_reset = bags_totals.reset_index()

# Convert the data from wide format to long format for easier plotting with seaborn
melted_bags_totals = pd.melt(bags_totals_reset, id_vars=['type'], var_name='Bag Size', value_name='Number of Bags')

# Create a bar plot using seaborn to visualize the total number of bags by avocado type and bag size
sns.barplot(x='type', y='Number of Bags', hue='Bag Size', data=melted_bags_totals)

# Set the plot title and axis labels with appropriate spacing
plt.title('Total Number of Small, Large, and XLarge Bags by Avocado Type', pad=10)
plt.xlabel('Avocado Type')
plt.ylabel('Number of Bags')

# Display the plot
plt.show()
```

##### 3. Analyze the avocado data from 2016 to 2018, to identify the top six regions where the sum of the average price of avocados is the highest. Create a visualization containing a grid of subplots, where each subplot is a box plot showing the distribution of average prices of organic and conventional avocados from 2016 to 2018, for the top six regions identified.

```python
# Filter the data to include only entries from the years 2016 to 2018
filtered_avocado_data = data[(data['year'] >= 2016) & (data['year'] <= 2018)]

# Group the filtered data by region and compute the sum of average prices for each region
region_price_sum = filtered_avocado_data.groupby('region')['AveragePrice'].sum().reset_index()
region_price_sum.rename(columns={'AveragePrice': 'TotalAveragePrice'}, inplace=True)

# Identify the top 6 regions with the highest total sum of average prices
top_six_regions = region_price_sum.nlargest(6, 'TotalAveragePrice')['region']

# Filter the dataset to include only the top 6 regions
top_region_data = data[data['region'].isin(top_six_regions)]

# Create separate DataFrames for organic and conventional avocados within the top 6 regions
organic_avocados = top_region_data[top_region_data['type'] == 'organic']
conventional_avocados = top_region_data[top_region_data['type'] == 'conventional']

# Create a 3x2 grid of subplots for displaying box plots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 9))
axes = axes.flatten()

# Generate a box plot for each of the top 6 regions
for idx, region in enumerate(top_six_regions):
    if idx < len(axes):
        organic_prices = organic_avocados[organic_avocados['region'] == region]['AveragePrice']
        conventional_prices = conventional_avocados[conventional_avocados['region'] == region]['AveragePrice']
        
        price_comparison = pd.DataFrame({
            'Organic': organic_prices,
            'Conventional': conventional_prices
        })
        
        price_comparison.plot(kind='box', ax=axes[idx], title=region)
        axes[idx].set_ylabel('Price')
        
# Adjust the layout to prevent overlap and display the plot
plt.tight_layout()
plt.show()
```
